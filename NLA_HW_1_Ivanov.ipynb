\begin{equation}
 \text{For } \|\cdot\|^{'} \text{ we need it to meet all 3 qualities of a norm:}

<br>$\text{1) } \|x\|^{'} \geq 0 \text{, and } \|x\|^{'} = 0, \text{only if } x = 0$

<br>\begin{equation} \|x\|^{'} = \sup\left\{|y^{\star} x|:\|y\| = 1\right\} \text{, so because of module propertires } |\cdot| \geq 0 \text{ and prescription that } ||y|| = 1 \text{, so }\end{equation}

<br>
<br>\begin{equation} \|\alpha x\|^{'} = \sup\left\{|y^{\star} \alpha x| : \|y\| = 1\right\} = \sup\left\{|\alpha||y^{\star} x| : \|y\| = 1\right\} = |\alpha|\sup\left\{|y^{\star} x| : \|y\| = 1\right\} = |\alpha|\cdot\|x\|^{'}

<br>$\text{3) } \|x + z\|^{'} \leq \|x\|^{'} + \|z\|^{'}$
<br>
<br>\begin{equation} \|x + z\|^{'} = \sup\left\{|y^{\star} (x + z)| : \|y\| = 1\right\} = \sup\left\{|y^{\star}x + y^{\star}z| : \|y\| = 1\right\} \text{ by module properties } \end{equation}
\begin{equation}\sup\left\{|y^{\star}x + y^{\star}z| : \|y\| = 1\right\} \leq \sup\left\{|y^{\star}x| : \|y\| = 1\right\} + \sup\left\{|y^{\star}z| : \|y\| = 1\right\} \leq \|x\|^{'} + \|z\|^{'}

```python
%%time
import numpy as np
import random
cos_rank, exp_rank, cos_exp_rank = [], [], []
r = 0
while r < 3:
    x_samples, y_samples = random.randint(0, 500), random.randint(0, 500)
    x, y = np.random.uniform(0, 1, x_samples), np.random.uniform(0, 1, y_samples)
    cos = np.zeros((x_samples, y_samples))
    exp = np.zeros((x_samples, y_samples))
    cos_exp = np.zeros((x_samples, y_samples))
    for i in range(x_samples):
        for j in range(y_samples):
            cos[i][j] = np.cos(x[i]*y[j])
            exp[i][j] = np.exp(-(x[i]-y[j])**2)
            cos_exp[i][j] = np.cos(x[i]*y[j]) * np.exp(-(x[i]-y[j])**2)
    
    # Singular Value Decomposition
    U, S_cos, V = np.linalg.svd(cos)
    U, S_exp, V = np.linalg.svd(exp)
    U, S_cos_exp, V = np.linalg.svd(cos_exp)
    
    cos_rank.append(len(S_cos[S_cos > 1e-5]))
    exp_rank.append(len(S_exp[S_exp > 1e-5]))
    cos_exp_rank.append(len(S_cos_exp[S_cos_exp > 1e-5]))
    r += 1
print(cos_rank, exp_rank, cos_exp_rank)
```

+ (1 pts). [RBF kernel](https://scikit-learn.org/stable/modules/gaussian_process.html#radial-basis-function-rbf-kernel) for different $N$.
\begin{equation}\text{RBF kernel: $k(x_i, y_j) = exp(- \frac{d(x_i, y_j)^{2}}{2l^{2}})$}\end{equation}

```python
%%time
def RBF(x, y):
    return np.exp(-(np.sqrt(np.sum((x-y)**2)))**2)             
    
N = 1    
while N < 10:
    rank_rbf, r = [], 0
    while r < 3:
        x_samples, y_samples = random.randint(0, 500), random.randint(0, 500)
        x, y = [], []
        for p in range(x_samples):
            x.append(np.random.uniform(0, 1, N))
        for q in range(y_samples):
            y.append(np.random.uniform(0, 1, N))

        rbf = np.zeros((x_samples,y_samples))
        for i in range(x_samples):
            for j in range(y_samples):
                rbf[i][j] = RBF(x[i], y[j])
                
        U, S_RBF, V = np.linalg.svd(rbf)
        rank_rbf.append(len(S_RBF[S_RBF > 1e-5]))
        r += 1

    print(f'N: {N}, RBF: {rank_rbf}')
    N += 2
```

+ (7 pts). Learn about [Kabsch algorithm](https://en.wikipedia.org/wiki/Kabsch_algorithm). Implement it and run on the database. Produce a list of distinct molecules and visualise them.

\begin{equation}\text{The Kabsch algorithm is a method for calculating the optimal rotation matrix that minimizes the RMSD (root mean squared deviation) between two paired sets of points}\end{equation}

```python
def Kabsch_algo(P, Q):
    """
    The algorithm for the rotation of P into Q starts with two sets of paired points, P and Q
    """

    # subtracting from the point coordinates of the respective centroid.
    P = (P - P.mean(0, keepdims = True))
    Q = (Q - Q.mean(0, keepdims = True))

    # Computation of the covariance matrix
    H = jnp.matmul(P.T, Q) 

    # Computation of the optimal rotation matrix
    # 1) calculate the SVD of the covariance matrix
    U, Sigma, VT = jnp.linalg.svd(H)

    # 2) Next, decide whether we need to correct our rotation matrix 
    # to ensure a right-handed coordinate system
    d = math.copysign(1.0, jnp.linalg.det(jnp.matmul(U, VT)))
    
    # 3) calcultation of Identity matrix
    n = P.shape[1]
    I = jnp.eye(n, dtype = jnp.float32).at[n-1, n-1].set(d)

    #Finally, calculate our optimal rotation matrix, R, as
    R = jnp.matmul(jnp.matmul(VT.T, I), U.T)
    
    return R
```

```python
%%time

group_molecules, molecules_by_group = {}, {}
for var1 in range(molecules.shape[0]):
    for var2 in range(molecules.shape[0]):
        P, Q = molecules[var1], molecules[var2]
        dcit1_var = molecules_by_group.get(var1) 
        dict2_var = molecules_by_group.get(var2)    
    
        R = Kabsch_algo(P, Q)
        if jnp.allclose(P @ R.T, Q):  # using the Kabsch algo to check
            if (dcit1_var is None) and (dict2_var is None):
                group_number = len(group_molecules) + 1
                group_molecules[group_number] = {var1, var2}
                molecules_by_group[var1] = group_number
                molecules_by_group[var2] = group_number
                continue
            if (dcit1_var is not None) and (dict2_var is not None):
                if dcit1_var != dict2_var:
                    raise ValueError

            if dcit1_var is None:
                number_in_list = var1 
            else:
                number_in_list = var2
            if dict2_var is None: 
                group_number = dcit1_var 
            else:
                group_number = dict2_var
            group_molecules[group_number].add(number_in_list)
            molecules_by_group[number_in_list] = group_number
```

# Problem 4 (Disentangled representation) (12 pts)

## Brief intro

The main component of [GAN](https://pytorch.org/tutorials/beginner/dcgan_faces_tutorial.html) is Generator network. Generator is such model that takes as input random vector and produces an image. In this task, we will work with the already pre-trained generator $G$. Here are a few things you need to know about this model:



1.   **Model input:** batch of random vectors $z\sim N(0,1)$ with normally distributed values of size **(Batch_Size,100)**. Here, Batch_Size is a number of images we want to generate
2.   **Model output:** $\mathrm{images} = G(z)$ - batch of images with size **(Batch_Size,3,64,64)**. Here **3** - defines number of image channels (3 for RGB images), **64, 64** - width and height.
3.   **Model structure:** generator is a set of sequentially appplied functions. Each function input is the output of the previous one. In other words, $G(z) = f_N(f_{N-1}...f_1(f_0(z)))$. $N$ - defines number of functions in generator.



### Disentanglement

The good property of GAN is that generator input space has a set of meaningful directions - moving input vector along this directions produce interpretable changes in image. For example, we can change a Pose, Age, Expression or even add eyeglasses. You can read more about it on [resource](https://genforce.github.io/interfacegan/). The main question is how to find such good directions. This problem is called disentanglement.

![](https://github.com/genforce/interfacegan/blob/master/docs/assets/teaser.jpg?raw=true)

## Problem discription 

We are going to find meaningful directions using SVD applied over the Jacobian matrix of some hidden output with respect to model input. **i-th** hidden output of generator model is defined by formula $G_{i}(z) = f_i(f_{i-1}...f_1(f_0(z))))$, where $i < N$. So, the task could be splitted on four main parts:

1. **(4 pts)**   Write the function **Get_Hidden_Output**, which reproduces formula $G_{i}(z) = f_i(f_{i-1}...f_1(f_0(z))))$
2. **(3 pts)**   Calculate jacobian matrix $J$ of generator $G$ for several **i** values. For each of them:

*   Calculate the jacobian matrix with respect to $z$: $\frac{\partial G(z)}{\partial z}$
*   Reshape jacobian matrix to have the size **(Batch_Size * Hidden_Output_Size,100)**

3. **(3 pts)**  Apply SVD to jacobian matrix ($J = U @ S @ V.T$) from previous step

4. **(2 pts)**  Take several directions from matrix $V$ (particular columns), and run visualization code (already written for you). Leave a comment on what you got (how found directions affect image)

```python
class Generator(nn.Module):
    def __init__(self, ngpu):
        super(Generator, self).__init__()
        self.ngpu = ngpu
        self.main = nn.Sequential(
            # input is Z, going into a convolution
            nn.ConvTranspose2d( nz, ngf * 8, 4, 1, 0, bias=False),
            nn.BatchNorm2d(ngf * 8),
            nn.ReLU(True),
            # state size. (ngf*8) x 4 x 4
            nn.ConvTranspose2d(ngf * 8, ngf * 4, 4, 2, 1, bias=False),
            nn.BatchNorm2d(ngf * 4),
            nn.ReLU(True),
            # state size. (ngf*4) x 8 x 8
            nn.ConvTranspose2d( ngf * 4, ngf * 2, 4, 2, 1, bias=False),
            nn.BatchNorm2d(ngf * 2),
            nn.ReLU(True),
            # state size. (ngf*2) x 16 x 16
            nn.ConvTranspose2d( ngf * 2, ngf, 4, 2, 1, bias=False),
            nn.BatchNorm2d(ngf),
            nn.ReLU(True),
            # state size. (ngf) x 32 x 32
            nn.ConvTranspose2d( ngf, nc, 4, 2, 1, bias=False),
            nn.Tanh()
            # state size. (nc) x 64 x 64
        )
    def get_function_output(self, inp, index=0):
        if index == 0:
            inp = inp[:,:,None,None]
        return self.main[index](inp)
    def get_number_of_layers(self):
        return len(self.main)

    def forward(self, input):
        return self.main(input[:,:,None,None])
```
